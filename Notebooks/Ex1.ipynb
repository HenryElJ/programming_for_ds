## 2. Symbol Counter

```python
def counter(*argv):
    arg_list = [x for x in argv]
    chars = list("".join(arg_list))
    chars_count = sorted([(x, chars.count(x)) for x in set(chars)])
    word_count = list(set([x for x in arg_list if arg_list.count(x) > 1]))

    print(chars_count)
    print(word_count)


counter("testing", "this", "script", "to", "see", "if", "this", "script", "works", "...?")
```

## 3. Sieve of Eratosthenes

```python
def sieve(n):
    numbers = [i for i in range(2, n + 1)]
    primes = []
    while len(numbers) != 0:
        primes.append(numbers[0])
        numbers = [numbers[j] for j in range(len(numbers)) if [i % numbers[0] != 0 for i in numbers][j]]
    
    return primes
```

```python
sieve(500)
```
