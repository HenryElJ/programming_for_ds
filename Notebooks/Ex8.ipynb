## 1. NumPy masking

```python
import numpy as np
import pandas as pd
```

```python
years = np.random.randint(0, 5000, 100)
```

```python
leap_years = years[[i % 4 == 0 and (i % 100 != 0 or i % 400 == 0) for i in years]]
```

```python
len(leap_years)
min(leap_years); max(leap_years)
np.sort(leap_years)[::-1]
```
